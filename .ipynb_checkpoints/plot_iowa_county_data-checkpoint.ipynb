```python

    
import geopandas as gp
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
from mpl_toolkits.axes_grid1 import make_axes_locatable
from datetime import datetime
import math    
import seaborn as sns
from matplotlib.colors import ListedColormap
    
```

```python
# turnout data
turnout_data_path = 'data/iowa_counties_turnout.json'
cities_data_path = 'data/iowa_cities_registered.json'

df_turnout = pd.read_json(turnout_data_path)
print(df_turnout.head())

df_cities = pd.read_json(cities_data_path)
print(df_cities.head())
 # todo match the decimal point convention on the shape file lat/long coordinates

# map - county shapes data 
data_path = 'data/shapefiles/iowa counties/cleaned_counties.shp'
shapes = gp.GeoDataFrame.from_file(data_path)
identifier_column = 'COUNTY' # name of column where you can find county names
print(shapes.head())

```

```python
# helper function for sizing city dots

def pop2radius(population): # scale area with population
    scalar = 15
    radius = math.sqrt( 2*population/math.pi ) # from the equation for area of a circle
    return radius*scalar
```

```python
# try plot the cities


plt.figure()
fig, ax = plt.subplots()

for idx, city in df_cities.iterrows(): # plot the cities
    population = city['Estimate']
    r = pop2radius(population)
    (x,y) = (city['MapX'],city['MapY'])
    #plt.plot(x,y,'x',alpha=0.5,linewidth=1)
    ax.add_patch(plt.Circle((x,y),r,color='#111111',alpha=0.25))

for idx, row in shapes.iterrows():  # plot the counties
    county_name = shapes.loc[idx,'COUNTY']
    shapes.loc[[idx],'geometry'].plot(ax=ax,alpha=0.25,linewidth=1,edgecolor='black',facecolor='white')

#for idx in range(shapes.shape[0]): # counties
#    county_name = shapes.loc[idx,identifier_column]
#    shapes.loc[[idx],'geometry'].plot(ax=ax, alpha=1, linewidth=0, edgecolor='black',facecolor=plot_color)
#plot.axis('off')
plt.title('test cities superposition')




```

```python
# voter turnout

plt.figure()
fig, ax = plt.subplots()
#ax = plt.gca()

cmap = plt.get_cmap('Blues')


    
for idx in range(shapes.shape[0]):
    county_name = shapes.loc[idx,identifier_column]    

    ballotsCast = df_turnout.loc[df_turnout['County']==county_name]['BallotsCast']
    registeredVoters = df_turnout.loc[df_turnout['County']==county_name]['TotalVoters']
        
    color_quantity = (1.0*ballotsCast) / registeredVoters
    plot_color = cmap(color_quantity)

    shapes.loc[[idx],'geometry'].plot(ax=ax, alpha=1, linewidth=0, edgecolor='black',facecolor=plot_color)
    
for idx, city in df_cities.iterrows(): # plot the cities
    population = city['Estimate']
    r = pop2radius(population)
    (x,y) = (city['MapX'],city['MapY'])
    #plt.plot(x,y,'x',alpha=0.5,linewidth=1)
    ax.add_patch(plt.Circle((x,y),r,edgecolor='#000000',alpha=1,fill=False,linewidth=0.5))
    


plt.axis('off')
plt.title('voter turnout')

#fig.subplots_adjust(bottom=0.5)
the_divider = make_axes_locatable(ax)
color_axis = the_divider.append_axes("right", size="5%", pad=0.1)
norm = mpl.colors.Normalize(vmin=0, vmax=1)
cb1 = mpl.colorbar.ColorbarBase(color_axis, cmap=cmap,
                                norm=norm,
                                orientation='vertical')
cb1.set_label('Voter Turnout Fraction')

plt.savefig('figures/iowa_county_turnout.png')
plt.show()
```

```python
# vote percentage

plt.figure()
ax = plt.gca()

ballots_cast = df_turnout['BallotsCast']
max_ballots = np.max(ballots_cast)
cmap = plt.get_cmap('Blues')

for idx in range(shapes.shape[0]):
    county_name = shapes.loc[idx,identifier_column]

    ballotsCast = df_turnout.loc[df_turnout['County']==county_name]['BallotsCast']
    #registeredVoters = df.loc[df['County']==county_name]['TotalVoters']
    
    color_quantity = (1.0*ballotsCast) / max_ballots
    plot_color = cmap(color_quantity)
    shapes.loc[[idx],'geometry'].plot(ax=ax, alpha=1, linewidth=0.1, edgecolor='navy',facecolor=plot_color)
    
for idx, city in df_cities.iterrows(): # plot the cities
    population = city['Estimate']
    r = pop2radius(population)
    (x,y) = (city['MapX'],city['MapY'])
    #plt.plot(x,y,'x',alpha=0.5,linewidth=1)
    ax.add_patch(plt.Circle((x,y),r,edgecolor='#000000',alpha=1,fill=False,linewidth=0.5))
    
plt.axis('off')
plt.title('Total Votes In Iowa Counties: 2018 Election')

the_divider = make_axes_locatable(ax)
color_axis = the_divider.append_axes("bottom", size="5%", pad=0.1)
norm = mpl.colors.Normalize(vmin=0, vmax=max_ballots)
cb1 = mpl.colorbar.ColorbarBase(color_axis, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')
cb1.set_label('total ballots cast')
plt.savefig('figures/total_ballots.png')
plt.show()
```

```python

sorted_vote_tallies = np.sort(ballots_cast)[::-1]
sorted_vote_percent = sorted_vote_tallies / np.sum(sorted_vote_tallies)

print(sorted_vote_percent)

plt.figure()
plt.plot(np.cumsum(sorted_vote_percent))
plt.xlabel('county')
plt.ylabel('cumulative fraction')

plt.savefig('figures/cumulative_vote_fraction.png')
plt.show()
```

```python

governor_data_path = 'data/iowa_counties_governor.json'

df_gov = pd.read_json(governor_data_path)
df_gov.head()
```

```python
# plot difference between governor's and congressional


# vote percentage

plt.figure()


ax = plt.gca()

vote_difference = df_gov['VoteDifference']
max_difference = np.max(vote_difference)
min_difference = -max_difference # np.min(vote_difference)
#cmap = plt.get_cmap('bwr')
#cmap = ListedColormap(sns.color_palette('RdBu',3000).as_hex())
cmap = ListedColormap(sns.color_palette('coolwarm',1000).as_hex()[::-1])

for idx, city in df_cities.iterrows(): # plot the cities
    population = city['Estimate']
    r = pop2radius(population)
    (x,y) = (city['MapX'],city['MapY'])
    #plt.plot(x,y,'x',alpha=0.5,linewidth=1)
    ax.add_patch(plt.Circle((x,y),r,color='#111111',alpha=0.25)) # todo change the style
    
for idx in range(shapes.shape[0]):
    county_name = shapes.loc[idx,identifier_column]

    vote_difference = df_gov.loc[df_gov['County']==county_name]['VoteDifference']
    
    color_quantity = (1.0*vote_difference - min_difference) / (max_difference - min_difference)
    #print(color_quantity)
    plot_color = cmap(color_quantity)
    shapes.loc[[idx],'geometry'].plot(ax=ax, alpha=1, linewidth=0.5, edgecolor='grey',facecolor=plot_color)
    
plt.axis('off')
csfont = {'fontname':'Roboto Black'}
plt.title('Vote Difference In Iowa Counties: 2018 Election (Governor)',**csfont)

the_divider = make_axes_locatable(ax)
color_axis = the_divider.append_axes("bottom", size="5%", pad=0.1)
norm = mpl.colors.Normalize(vmin=min_difference, vmax=max_difference)
cb1 = mpl.colorbar.ColorbarBase(color_axis, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')


hfont = {'fontname':'Roboto Black'}
cb1.set_label('vote differential',**hfont)
plt.savefig('figures/vote_differential.png')



plt.show()

```

```python
# democrats: US House votes - governor votes

plt.figure()
ax = plt.gca()

net_difference = df_house['DemTotal'] - df_gov['HubbellTotal']
net_difference_dem = net_difference # rename for convenience
#max_net_difference = -min_net_difference # from inspection np.max(net_difference)
max_net_difference = np.max(net_difference)
min_net_difference = - max_net_difference # from inspection, for color axis equality
#cmap = plt.get_cmap('bwr')
#cmap = ListedColormap(sns.color_palette('RdBu',3000).as_hex())
#cmap = ListedColormap(sns.color_palette('coolwarm',1000).as_hex()[::-1])
cmap = ListedColormap(sns.color_palette('PuOr',30).as_hex())
#cmap = ListedColormap(sns.palplot(sns.diverging_palette(150, 275, s=80, l=55, n=9)))


    
# paint in the counties
for idx in range(shapes.shape[0]):
    county_name = shapes.loc[idx,identifier_column]

    this_difference =  df_house.loc[df_house['County']==county_name]['DemTotal']  -      \
                           df_gov.loc[df_gov['County']==county_name]['HubbellTotal']
    
    color_quantity = (1.0*this_difference - min_net_difference) / (max_net_difference - min_net_difference)
    #print(color_quantity)
    plot_color = cmap(color_quantity)
    shapes.loc[[idx],'geometry'].plot(ax=ax, alpha=1, linewidth=0.5, edgecolor='grey',facecolor=plot_color)
    
# paint in the cities
for idx, city in df_cities.iterrows(): # plot the cities
    population = city['Estimate']
    r = pop2radius(population)
    (x,y) = (city['MapX'],city['MapY'])
    #plt.plot(x,y,'x',alpha=0.5,linewidth=1)
    ax.add_patch(plt.Circle((x,y),r,color='#111111',alpha=0.25)) # todo change the style
    
plt.axis('off')
csfont = {'fontname':'Roboto Black'}
plt.title('Democratic Vote Difference: Congressional Races vs Governor',**csfont)

the_divider = make_axes_locatable(ax)
color_axis = the_divider.append_axes("bottom", size="5%", pad=0.1)
norm = mpl.colors.Normalize(vmin=min_net_difference, vmax=max_net_difference)
cb1 = mpl.colorbar.ColorbarBase(color_axis, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')


hfont = {'fontname':'Roboto Black'}
cb1.set_label('D vote differential',**hfont)
plt.savefig('figures/dems_house_minus_gov.png')



plt.show()



```

```python
# normalize by ballots cast

# democrats: US House votes - governor votes

plt.figure()
ax = plt.gca()

net_difference = (df_house['DemTotal'] - df_gov['HubbellTotal']) / ballots_cast
net_difference_dem = net_difference # rename for convenience
#max_net_difference = -min_net_difference # from inspection np.max(net_difference)
max_net_difference = np.max(net_difference)
min_net_difference = - max_net_difference # from inspection, for color axis equality
#cmap = plt.get_cmap('bwr')
#cmap = ListedColormap(sns.color_palette('RdBu',3000).as_hex())
#cmap = ListedColormap(sns.color_palette('coolwarm',1000).as_hex()[::-1])
cmap = ListedColormap(sns.color_palette('PuOr',30).as_hex())
#cmap = ListedColormap(sns.palplot(sns.diverging_palette(150, 275, s=80, l=55, n=9)))


    
# paint in the counties
for idx in range(shapes.shape[0]):
    county_name = shapes.loc[idx,identifier_column]

    this_difference =  (df_house.loc[df_house['County']==county_name]['DemTotal']  -      \
                           df_gov.loc[df_gov['County']==county_name]['HubbellTotal']) /    \
                            df_turnout[ df_turnout['County']==county_name]['BallotsCast']
    
    
    color_quantity = (1.0*this_difference - min_net_difference) / (max_net_difference - min_net_difference)
    #print(color_quantity)
    plot_color = cmap(color_quantity)
    shapes.loc[[idx],'geometry'].plot(ax=ax, alpha=1, linewidth=0.5, edgecolor='#ffffff',facecolor=plot_color)
    
# paint in the cities
for idx, city in df_cities.iterrows(): # plot the cities
    population = city['Estimate']
    r = pop2radius(population)
    (x,y) = (city['MapX'],city['MapY'])
    #plt.plot(x,y,'x',alpha=0.5,linewidth=1)
    ax.add_patch(plt.Circle((x,y),r,color='#5a9aac',alpha=0.75)) # todo change the style
    
plt.axis('off')
csfont = {'fontname':'Roboto Black'}
plt.title('Normalized Vote Difference: Democratic Congressional Races - Democratic Governor',**csfont)

the_divider = make_axes_locatable(ax)
color_axis = the_divider.append_axes("bottom", size="5%", pad=0.1)
norm = mpl.colors.Normalize(vmin=min_net_difference, vmax=max_net_difference)
cb1 = mpl.colorbar.ColorbarBase(color_axis, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')


hfont = {'fontname':'Roboto Black'}
cb1.set_label('fraction of total vote',**hfont)
plt.savefig('figures/dems_house_minus_gov.png')



plt.show()

```

```python
# republican US House votes - governor votes

plt.figure()
ax = plt.gca()

net_difference = df_gov['ReynoldsTotal'] - df_house['RepubTotal']
net_difference_rep = net_difference # rename for convenience
max_net_difference = np.max(net_difference)
min_net_difference = -max_difference # # this gives us  a symmetric color axis
cmap = ListedColormap(sns.color_palette('PuOr',30).as_hex()[::-1])

# paint in the cities
for idx, city in df_cities.iterrows(): # plot the cities
    population = city['Estimate']
    r = pop2radius(population)
    (x,y) = (city['MapX'],city['MapY'])
    #plt.plot(x,y,'x',alpha=0.5,linewidth=1)
    ax.add_patch(plt.Circle((x,y),r,color='#111111',alpha=0.25)) # todo change the style
    
# paint in the counties
for idx in range(shapes.shape[0]):
    county_name = shapes.loc[idx,identifier_column]

    this_difference =  df_gov.loc[df_gov['County']==county_name]['ReynoldsTotal']  -      \
                          df_house.loc[df_house['County']==county_name]['RepubTotal']
    
    color_quantity = (1.0*this_difference - min_net_difference) / (max_net_difference - min_net_difference)
    #print(color_quantity)
    plot_color = cmap(color_quantity)
    shapes.loc[[idx],'geometry'].plot(ax=ax, alpha=1, linewidth=0.5, edgecolor='grey',facecolor=plot_color)
    
plt.axis('off')
csfont = {'fontname':'Roboto Black'}
plt.title('Republicans: Gov - US House',**csfont)

the_divider = make_axes_locatable(ax)
color_axis = the_divider.append_axes("bottom", size="5%", pad=0.1)
norm = mpl.colors.Normalize(vmin=min_net_difference, vmax=max_net_difference)
cb1 = mpl.colorbar.ColorbarBase(color_axis, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')


hfont = {'fontname':'Roboto Black'}
cb1.set_label('vote differential',**hfont)
plt.savefig('figures/GOP_vote_differential.png')



plt.show()
```
