```python
# 1. plot voter turnout

# 2. plot general election difference statistics

# 3.  plot primary differences

# 4.  correlate primary w general election

#import xml.etree.ElementTree as ElementTree  # parse xml
from bs4 import BeautifulSoup
import pandas as pd
#import json
import numpy as np

```

```python
# load data

data_path = 'data/detail_iowa_counties_general2018.xml' # use the xml data file, that should be easier
infile = open(data_path,"r")
contents = infile.read()
soup = BeautifulSoup(contents,'xml')

target_file_path = 'data/iowa_counties_turnout.json'
```

```python
# voter turnout by county

iowa_turnout = {'County':"Overall",
                'TotalVoters':soup.ElectionVoterTurnout['totalVoters'],
                'BallotsCast':soup.ElectionVoterTurnout['ballotsCast'],
                'VoterTurnout':soup.ElectionVoterTurnout['voterTurnout']}
print(iowa_turnout)

column_names = ['County','TotalVoters','BallotsCast','%']
#df = pd.DataFrame(columns=column_names)

table = soup.ElectionVoterTurnout.Counties.contents[0:]
data_objects = []
#print(table)
for idx,entry in enumerate(table):
    if (idx)%2 == 1: # for even indices of contents[idx]  (I don't quite understand why - are there empty entries?)
    
        print(entry)
        
        county = entry['name']
        N_voters = entry['totalVoters']
        N_ballots = entry['ballotsCast']
        percent = entry['voterTurnout']
        
        data_object = {"County":county,"TotalVoters":N_voters,"BallotsCast":N_ballots,"%":percent}
        data_objects.append(data_object)
        
df = pd.DataFrame(data_objects)
df.head()
#soup.ElectionVoterTurnout.Counties
```

```python
# export processed data file to local directory

df.to_json(orient='records',path_or_buf = target_file_path) # note could probably do better formatting here  ,lines=True

#with open(target_file_path, 'w') as outfile:
#     json.dump(raw_json, outfile, sort_keys = True, indent = 4) #  ensure_ascii = False)
```

```python
# governor's race
list_of_counties = df['County']
N_counties = len(list_of_counties)
df_gov = pd.DataFrame(index=np.arange(0,N_counties), columns=['County',
                                                               'HubbellElectionDay',
                                                               'HubbellAbsentee',
                                                               'ReynoldsElectionDay',
                                                               'ReynoldsAbsentee'])
df_gov['County'] = list_of_counties

contests = soup.find_all('Contest')
for contest in contests:
    #contest_list.append(contest['text'])
    if 'Governor' in contest['text']:
        choices = contest.find_all('Choice')
        for choice in choices:
            
            candidate = choice['text']
            print(candidate)
            
            vote_types = choice.find_all('VoteType')
            for vote_type in vote_types:
                print(vote_type['name'])
                
                counties = vote_type.find_all('County')
                for county in counties:
                    county_name = county['name']
                    N_votes = county['votes']
                        
                    if vote_type['name']=='Election Day':
                        if 'Hubbell' in candidate:
                            df_gov.loc[df_gov['County']==county_name,'HubbellElectionDay'] = N_votes

                        elif 'Reynolds' in candidate:
                            df_gov.loc[df_gov['County']==county_name,'ReynoldsElectionDay'] = N_votes
                    
                    
                    elif vote_type['name']=='Absentee':
                        if 'Hubbell' in candidate:
                            df_gov.loc[df_gov['County']==county_name,'HubbellAbsentee'] = N_votes

                        elif 'Reynolds' in candidate:
                            df_gov.loc[df_gov['County']==county_name,'ReynoldsAbsentee'] = N_votes
                
            #election_day_data = choice.find_all(VoteType['name']=='Election Day')
            #print(election_day_data)
df_gov['HubbellTotal'] = [int(x) + int(y) for x,y in zip(df_gov['HubbellElectionDay'],df_gov['HubbellAbsentee'])]
df_gov['ReynoldsTotal'] = [int(x) + int(y) for x,y in zip(df_gov['ReynoldsElectionDay'],df_gov['ReynoldsAbsentee'])]
df_gov['VoteDifference'] = [int(x) - int(y) for x,y in zip(df_gov['HubbellTotal'],df_gov['ReynoldsTotal'])]
df_gov.head()
```

```python
gov_target_file_path = 'data/iowa_counties_governor.json'
df_gov.to_json(orient='records',path_or_buf = gov_target_file_path) # note could probably do better formatting here  ,lines=True
```

```python
# US house reps

    
df_UShouse = pd.DataFrame(index=np.arange(0,N_counties), columns=['County',
                                                               'DemCandidate',
                                                               'DemElectionDay',
                                                               'DemAbsentee',
                                                               'RepubCandidate',
                                                               'RepubElectionDay',
                                                               'RepubAbsentee'])

UShouse_string = ['United States Representative District 1',
 'United States Representative District 2',
 'United States Representative District 3',
 'United States Representative District 4']

print(UShouse_string)
df_UShouse['County'] = list_of_counties

contests = soup.find_all('Contest')
for contest in contests:
    
    if contest['text'] in UShouse_string:
        print(contest['text'])
        
        choices = contest.find_all('Choice')
        for choice in choices:
            
            if choice.has_attr('party'): # because the write-in choice has no party affiliation attribute
                if choice['party']=='DEM':
                    candidate = choice['text']

                    vote_types = choice.find_all('VoteType')
                    for vote_type in vote_types:
                        print(vote_type['name'])

                        counties = vote_type.find_all('County')
                        for county in counties:
                            county_name = county['name']
                            N_votes = county['votes']

                            df_UShouse.loc[df_UShouse['County']==county_name,'DemCandidate'] = candidate

                            if vote_type['name']=='Election Day':
                                df_UShouse.loc[df_UShouse['County']==county_name,'DemElectionDay'] = N_votes

                            elif vote_type['name']=='Absentee':
                                 df_UShouse.loc[df_UShouse['County']==county_name,'DemAbsentee'] = N_votes


                elif choice['party']=='REP':
                    candidate = choice['text']

                    vote_types = choice.find_all('VoteType')
                    for vote_type in vote_types:
                        print(vote_type['name'])

                        counties = vote_type.find_all('County')
                        for county in counties:
                            county_name = county['name']
                            N_votes = county['votes']
                            #print(county_name, N_votes, candidate)

                            df_UShouse.loc[df_UShouse['County']==county_name,'RepubCandidate'] = candidate

                            if vote_type['name']=='Election Day':
                                df_UShouse.loc[df_UShouse['County']==county_name,'RepubElectionDay'] = N_votes

                            elif vote_type['name']=='Absentee':
                                 df_UShouse.loc[df_UShouse['County']==county_name,'RepubAbsentee'] = N_votes

print(df_UShouse.head())                                
df_UShouse['DemTotal'] = [int(x) + int(y) for x,y in zip(df_UShouse['DemElectionDay'],df_UShouse['DemAbsentee'])]
df_UShouse['RepubTotal'] = [int(x) + int(y) for x,y in zip(df_UShouse['RepubElectionDay'],df_UShouse['RepubAbsentee'])]
df_UShouse['VoteDifference'] = [int(x) - int(y) for x,y in zip(df_UShouse['DemTotal'],df_UShouse['RepubTotal'])]
df_UShouse.head()
```
