```python
    
import geopandas as gp
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
from mpl_toolkits.axes_grid1 import make_axes_locatable



data_path = 'data/shapefiles/iowa counties/county.shp'
shapes = gp.GeoDataFrame.from_file(data_path)
identifier_column = 'COUNTY' # name of column where you can find county names
print(shapes.head())



```

```python
# HELPER FUNCTION

# find the nearest point
def get_nearest_idx(X, Y, x_coords, y_coords): # x target, y target, list of x coordinates, list of y coordinates
    coord_distances = np.abs(x_coords - X) + np.abs(y_coords - Y)
    nearest_idx = np.argmin(coord_distances)
    return nearest_idx
```
